```python
# Sentence Tokenization

### The first step imports NLTK and downloads the 'punkt' package for sentence tokenization. It then imports sent_tokenize and applies it to a variable named text. Finally, it iterates through the sentences and prints them with their corresponding index.  

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize


text = "Your sample corpus text goes here. It should have multiple sentences."

sentences = sent_tokenize(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print(tokens[:10]) 
print(len(tokens))
```

```python
# Lowercasing & Vocabulary Size

lower_tokens = [word.lower() for word in tokens]
print(lower_tokens[:20])

# Vocab size before
vocab_before = set(tokens)
print(len(vocab_before))

# Vocab size after
vocab_after = set(lower_tokens)
print(len(vocab_after))
```

```python
# Removing Punctuation

import string

tokens_no_punct = [
    word for word in lower_tokens 
    if word not in string.punctuation
]
print(tokens_no_punct)
```

```python
# Removing Stopwords

nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

filtered_tokens = [
    word for word in tokens_no_punct 
    if word not in stop_words
]
print(filtered_tokens)
```

```python
# Frequency Distribution

from collections import Counter

freq = Counter(filtered_tokens)
print(freq.most_common())
```

```python
# Stemming

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stems = [
    stemmer.stem(word) 
    for word in filtered_tokens
]
print(stems)
```

```python
# Lemmatization

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmas = [
    lemmatizer.lemmatize(word) 
    for word in filtered_tokens
]
print(lemmas)
```

```python
# Saving the Final Corpus

final_corpus = " ".join(lemmas)

with open("final_corpus.txt", "w", encoding="utf-8") as file:
    file.write(final_corpus)

print("File saved successfully")
```
