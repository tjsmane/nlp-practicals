```python
# Sentence Tokenization

### The first step imports NLTK and downloads the 'punkt' package for sentence tokenization. It then imports sent_tokenize and applies it to a variable named text. Finally, it iterates through the sentences and prints them with their corresponding index.  

import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# Assuming 'text' is defined earlier in your notebook as your corpus
text = "Your sample corpus text goes here. It should have multiple sentences."

sentences = sent_tokenize(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
```
